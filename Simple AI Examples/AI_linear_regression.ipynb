```python
import numpy as np
import matplotlib.pyplot as plt

# Входные данные
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# График входных данных
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Исходные данные')
plt.show()

# Вычисление среднего значения X и y
mean_x = np.mean(X)
mean_y = np.mean(y)

# Вычисление разностей для X и y
diff_x = X - mean_x
diff_y = y - mean_y

# Вычисление коэффициентов регрессии
slope = np.sum(diff_x * diff_y) / np.sum(diff_x**2)
intercept = mean_y - slope * mean_x

# Предсказание
x_new = 6
y_pred = slope * x_new + intercept

# График регрессии
plt.scatter(X, y)
plt.plot(X, slope * X + intercept, color='red')
plt.scatter(x_new, y_pred, color='green')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Модель линейной регрессии')
plt.legend(['Линия регрессии', 'Предсказанное значение'])
plt.show()

# Вывод результата
print("Предсказанное значение:", y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Загрузка данных из файла
data = pd.read_csv('mobile_phones.csv')
X = data['battery_power']
y = data['talk_time']

# График входных данных
plt.scatter(X, y)
plt.xlabel('battery_power')
plt.ylabel('talk_time')
plt.title('Исходные данные')
plt.show()

# Вычисление среднего значения X и y
mean_x = np.mean(X)
mean_y = np.mean(y)

# Вычисление разностей для X и y
diff_x = X - mean_x
diff_y = y - mean_y

# Вычисление коэффициентов регрессии
slope = np.sum(diff_x * diff_y) / np.sum(diff_x**2)
intercept = mean_y - slope * mean_x

# Предсказание
x_new = 2000
y_pred = slope * x_new + intercept

# График регрессии
plt.scatter(X, y)
plt.plot(X, slope * X + intercept, color='red')
plt.scatter(x_new, y_pred, color='green')
plt.xlabel('battery_power')
plt.ylabel('talk_time')
plt.title('Модель линейной регрессии')
plt.legend(['Линия регрессии', 'Предсказанное значение'])
plt.show()

# Вывод результата
print("Предсказанное значение:", y_pred)
```
